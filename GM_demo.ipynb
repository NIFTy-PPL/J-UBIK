```python
import nifty8.re as jft
import jax
from jax import random
import matplotlib.pyplot as plt
```

```python
key = jax.random.PRNGKey(42)
key, subkey = jax.random.split(key)
```

## WIENER PROCESS

```python
wp = jft.WienerProcess(0, 3, 2, "margret", 100)
wp.domain 
```

```python
pos_init = jft.Vector(jft.random_like(subkey, wp.domain))
field = wp(pos_init)
plt.plot(field)
plt.show()
plt.close()
```

# Integrated WP

```python
iwp = jft.IntegratedWienerProcess(x0=(0,1), sigma=(1,3), dt=2, name="Vincent", N_steps=100)
```

```python
key, subkey = jax.random.split(key)
pos_init = jft.Vector(jft.random_like(subkey, iwp.domain))
field = iwp(pos_init)
plt.clf()
plt.plot(field)
plt.show()
plt.close()
```

# OrnsteinUhlenbeck

```python
ou = jft.OrnsteinUhlenbeckProcess(sigma=10, gamma=1, dt=1, name="matteo", x0=0, N_steps=100)
```

```python
key, subkey = jax.random.split(key)
pos_init = jft.Vector(jft.random_like(subkey, ou.domain))
field = ou(pos_init)
plt.clf()
plt.plot(field)
plt.show()
plt.close()
```
